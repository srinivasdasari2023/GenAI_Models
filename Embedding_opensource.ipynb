```python
from langchain.embeddings.huggingface import HuggingFaceBgeEmbeddings
from llama_index import ServiceContext
from llama_index.embeddings import LangchainEmbedding
```

```python
embed_model = LangchainEmbedding(
    HuggingFaceBgeEmbeddings(model_name = "sentence-transformers/all-mpnet-base-v2")
)
```

```python
service_context = ServiceContext.from_defaults(
    chunk_size = 1024,
    llm = llm,
    embed_model = embed_model
)
```

```python
index = VectorStoreIndex.from_documents(documents, service_context= service_context)
```

```python
index.as_query_engine()
```

```python
query_engine=index.as_query_engine()
```

```python
query = "What is correlation?"
response = query_engine.query(query)
print(response)
```

```python
import google.generativeai as genai
genai.configure(api_key= GOOGLE_API_KEY)
```

```python
for model in genai.list_models():
  print(model)
```

```python
for model in genai.list_models():
  if "generateContent" in model.supported_generation_methods:
    print(model.name)
```

```python
def load_model(model_name):
  if model_name=="gemini-pro":
   llm=genai.GenerativeModel("gemini-pro")
  else:
    #image text
    llm=genai.GenerativeModel("gemini-pro-vision")
  return llm
```

```python
model_text=load_model("gemini-pro")
```

```python
model_text
```

```python
model_vision=load_model("gemini-pro-vision")
```

```python
model_vision
```

```python
response = model_text.generate_content("what is the capital of india?")
```
